```python
import pyiges
iges = pyiges.read("Disk.iges")
```

```python
lines=iges.lines()
points=iges.points()
```

```python
import numpy
xl = numpy.array([line.coordinates[0][0] for line in lines])
yl = numpy.array([line.coordinates[0][1] for line in lines])
```

```python
xp = numpy.array([point.x for point in points])
yp = numpy.array([point.y for point in points])
```

```python
import matplotlib.pyplot as pyplot
pyplot.figure()
pyplot.plot(xl[:],yl[:])
#pyplot.scatter(xp,yp)
pyplot.axis('equal')
```

```python
theta = numpy.linspace(0, 2*numpy.pi, 10)
xcircle = numpy.cos(theta)
ycircle = numpy.sin(theta)

xA, yA = xcircle[1:], ycircle[1:]
xB, yB = xcircle[:-1], ycircle[:-1]

ny = xB-xA
nx = -(yB-yA)

tx = xB-xA
ty = yB-yA

xC = (xA+xB)/2
yC = (yA+yB)/2
```

```python
pyplot.figure()
pyplot.plot(xA[:],yA[:],'r')
pyplot.plot(xB[:],yB[:],'b')
pyplot.scatter(xC,yC)
pyplot.quiver(xC,yC,nx,ny)
pyplot.axis('equal')
```

```python
pyplot.figure()
pyplot.scatter(xA[0],yA[0], color='teal', marker='s', s=100)
pyplot.scatter(xB[0],yB[0], color='orange', marker='s', s=100)
pyplot.plot([xA[0], xB[0]],[yA[0], yB[0]],'k')
pyplot.scatter(xC[0],yC[0], color='k', marker='o', s=100)
pyplot.quiver(xC[0],yC[0],nx[0],ny[0], scale=5, color='b')
pyplot.quiver(xC[0],yC[0],tx[0],ty[0], scale=5, color='orange')
pyplot.axis('equal')
```
