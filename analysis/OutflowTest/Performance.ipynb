```python
import matplotlib.pyplot as pyplot
import matplotlib.ticker as mticker
```

```python
nodes = [1,5,10]
amrex = [0.09249550072137236, 0.14271265598738747, 0.23972602739726057]
paramesh = [1.1769517945056744, 1.8550414231106749, 3.1392879984531405]
```

```python
pyplot.rc("font", family="serif", size=12, weight="bold")
pyplot.rc("axes", labelweight="bold", titleweight="bold")
pyplot.rc("text", usetex=True)
figure =  pyplot.figure(figsize=(4,3), dpi=200)

pyplot.plot(
    nodes,
    paramesh,
    marker='s',
    markersize=9,
)

pyplot.plot(
    nodes,
    amrex,
    marker='o',
    markersize=9,
)

pyplot.ylabel(r"Average time per rank (s)")
pyplot.xlabel(r"Nodes (65 ranks/node)")
#pyplot.xscale('log')
pyplot.yscale('log')
pyplot.ylim([0.05,5])
#pyplot.grid()
#pyplot.grid(which="minor")
ax=pyplot.gca()
ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
ax.set_yticks([0.05,0.1,0.5,1.0,1,5])
pyplot.legend(["Flash-X with HYPRE","Flash-X with AMReX Multigrid"])
pyplot.tight_layout()
pyplot.savefig("/Users/Akash/Desktop/performance.png")
```
