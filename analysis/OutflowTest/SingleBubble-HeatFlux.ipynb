```python
import os
import sys
import json
import itertools
from scipy import signal
import numpy
import matplotlib.pyplot as pyplot
```

```python
heatFluxDict = json.load(open("SingleBubble-HeatFlux.json") )
growthDict = json.load(open("SingleBubble-Growth.json"))
```

```python
pyplot.rc("font", family="serif", size=10, weight="bold")
pyplot.rc("axes", labelweight="bold", titleweight="bold")
pyplot.rc("text", usetex=True)
figure =  pyplot.figure(figsize=(4,2), dpi=200)

pyplot.plot(
    numpy.array(heatFluxDict["Buffer = 0.5"]["time"])*.5,
    signal.savgol_filter(numpy.array(heatFluxDict["Buffer = 0.5"]["qliq"]), 30, 3),
)

pyplot.plot(
    numpy.array(heatFluxDict["Buffer = 1.0"]["time"])*.5,
    signal.savgol_filter(numpy.array(heatFluxDict["Buffer = 1.0"]["qliq"]), 30, 3),
    linestyle='--'
)
pyplot.ylim([3,5])
pyplot.ylabel(r"${q}_w$", rotation=0, labelpad=10)
pyplot.xlabel(r"$t$", labelpad=-1)
pyplot.legend(["$l_b = 0.5$","$l_b = 1.0$"])
pyplot.tight_layout()
pyplot.savefig("/Users/Akash/Desktop/hf.png")
```

```python
pyplot.rc("font", family="serif", size=10, weight="bold")
pyplot.rc("axes", labelweight="bold", titleweight="bold")
pyplot.rc("text", usetex=True)
figure =  pyplot.figure(figsize=(4,2), dpi=200)

pyplot.plot(growthDict["$l_b = 0.5$"]["time"][:-2], growthDict["$l_b = 0.5$"]["diameter"][:-2], marker='o', markevery=3)
pyplot.plot(growthDict["$l_b = 1.0$"]["time"][:-2], growthDict["$l_b = 1.0$"]["diameter"][:-2], linestyle='--')

pyplot.ylim([0,2])
pyplot.ylabel(r"$D_{g}$", rotation=0, labelpad=10)
pyplot.xlabel(r"$t_c$", labelpad=-1)
pyplot.legend(["$l_b = 0.5$","$l_b = 1.0$"])
pyplot.tight_layout()
pyplot.savefig("/Users/Akash/Desktop/growth.png")
```
